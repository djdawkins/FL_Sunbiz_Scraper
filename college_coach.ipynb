```python
import pandas as pd
```

```python
school_coach_df = pd.read_csv("school_coach_table.csv")
school_coach_df.head()
```

```python
def rename_cols(df):
    df = df.rename(columns={ df.columns[0]: "coach", df.columns[1]: "coach_url" })
    return df
```

```python
# rename_cols()
```

```python
hc_df = school_coach_df[["Head coach", "hc_url"]]

oc_df = school_coach_df[["Offensive coordinator(s)", "oc_url"]]
dc_df = school_coach_df[["Defensive coordinator(s)", "dc_url"]]
sc_df = school_coach_df[["Special teams coordinator(s)", "sc_url"]]

coach_df_list = [hc_df, oc_df,dc_df, sc_df]
output_list = [rename_cols(i) for i in coach_df_list]

all_coach_df = pd.concat(output_list)
all_coach_df.head()
```

```python
# Filter for Nan's
nan_coach_df  = all_coach_df[all_coach_df.coach_url.isna()]

coach_df_filt  = all_coach_df[all_coach_df.coach_url.notna()]
coach_df_filt.head()
```

```python
dual_coach_df = coach_df_filt.copy()

single_coach_df = dual_coach_df[dual_coach_df.coach.str.split("/").str.len() < 2]
dual_coach_df = dual_coach_df[dual_coach_df.coach.str.split("/").str.len() > 1]

dual_coach_df.head()
```

```python
dual_coach_df_exp = dual_coach_df.copy()

dual_coach_df_exp["coach"] = dual_coach_df_exp.coach.str.split("/")
dual_coach_df_exp["coach_url"] = dual_coach_df_exp.coach_url.str.split("|")

dual_coach_df_exp = dual_coach_df_exp.explode(['coach', 'coach_url'])

dual_coach_df_exp.head()
```

```python
all_coach_df = pd.concat([single_coach_df, dual_coach_df_exp])
all_coach_df.head()
```

```python
coach_df_good_link = all_coach_df[~all_coach_df.coach_url.str.contains("redlink")]
coach_df_good_link.to_csv("all_coaches.csv")
coach_df_good_link

# coach_df_red_link = all_coach_df[all_coach_df.coach_url.str.contains("redlink")]
# coach_df_red_link
```
